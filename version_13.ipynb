### 1B. Voter ID Law Data

Next we wanted to track when voter ID laws were implemented in different states in the past 14 years, separating laws into two general categories of strictness (strict and non-strict). The definitions for these categories are fairly consistent across different sources, but our specific definition was taken from the National Conference of State Legislatures website:
    <ul>
    <li>Strict: Voters without acceptable identification can vote only on a provisional ballot and must return within some number of days to present ID. </li>
    <li>Non-strict: At least some voters without acceptable identification can cast a ballot with some additional actions, such as signing an affidavit confirming their identity. </li>
    </ul>
Note that this does not capture differences in photo ID/non-photo ID requirements (each of these types can also be either strict or non-strict) since this would severely limit data points for each category. 

Unfortunately, we were not able to scrape to collect this data since we could not find any standardized source detailing implementation of voter ID laws, so dates were verified individually. 

```python
#Resources used to determine dates include:
 #http://www.ncsl.org/research/elections-and-campaigns/voter-id.aspx
 #http://en.wikipedia.org/wiki/Voter_ID_laws_in_the_United_States
 #http://www.propublica.org/article/everything-youve-ever-wanted-to-know-about-voter-id-laws - definition of non-strict
 #ballotpedia.org/State_by_State_Voter_ID_Laws
 #http://www.ncsl.org/research/elections-and-campaigns/voter-id-history.aspx#Chart
 #http://thelensnola.org/2013/06/27/louisianas-voter-id-law-from-1997-eases-effects-of-supreme-court-decision/

#Legislative proceedings were used -- all accessed through National Conference of State Legislatures website (listed above).
#Most dates were cross-checked with several of the other sources above. 
#Some states (like NC) currently have no laws but will implement them soon. 

#create dataframe for states and law implementation
law_id=pd.DataFrame(columns=['State','non_strict_year','strict_year'])
law_id['State'] = state[1:52]

# entering dates of enactment for each state
law_id.ix[0,1] = '2003-06-24'                         # beginning June 3, 2014 stricter photo ID law in effect
law_id.ix[1,1] = '1980-01-01'                         # estimated day of enactment (year is accurate)
law_id.ix[2,2] = '2004-12-01'                         # Proposition approved in November 2004 - estimated day of enactment
law_id.ix[3,1] = '2000-01-01'                         # non-strict law enacted at some point before 2000
law_id.ix[5,1] = '2003-04-22'                         # approved April 22, 2003, as of 2013 voters can now vote by mail, so can be circumvented
law_id.ix[6,1] = '2001-01-01'                         # non-strict law enacted at some point before 2002 (estimated day)
law_id.ix[7,1] = '2001-01-01'                         # non-strict law enacted at some point before 2002 (estimated day)
law_id.ix[9,1] = '1977-01-01'                         # estimated day of enactment (year is accurate)
law_id.ix[10,[1,2]] = ['1997-01-01','2008-06-01']     # estimated day of enactment, known to affect general elections that year
law_id.ix[11,1] = '1970-01-01'                        # estimated day of enactment (year is correct)  
law_id.ix[12,1] = '2010-06-01'                        # effective after May primary but before general election (estimated day of enactment)
law_id.ix[14,2] = '2008-05-01'                        # effective sometime after Supreme Court case decision in April (estimated day)
law_id.ix[16,2] = '2012-01-01'                        # this date is actually exact - the law was passed in 2011
law_id.ix[17,1] = '2000-01-01'                        # estimated day of enactment, know to exist before 2002 elections
law_id.ix[18,1] = '1997-01-01'                        # estimated day of enactment, year is correct  
law_id.ix[22,1] = '2007-08-01'                        # upheld in court near July 20, 2007 (estimated day of enactment)
law_id.ix[24,2] = '2014-06-03'                        # approved in 2012, did not go into effect until 2014
law_id.ix[25,1] = '2002-08-28'                        # day of approval based on legislative proceedings
law_id.ix[26,1] = '2004-01-01'                        # effective day based on legislative proceedings
law_id.ix[29,1] = '2012-11-01'                        # in effect for Nov 2012 elections, a strict law is going into effect in 2015
law_id.ix[34,[1,2]] = ['2003-12-01','2013-06-01']     # estimated day of effect for non-strict law based on proceedings, strict law is more accurate
law_id.ix[35,2] = '2006-06-01'                        # effective day based on legislative proceedings
law_id.ix[36,1] = '2010-12-01'                        # estimated day of enactment, approved by voters in 2010, modified in 2013
law_id.ix[39,1] = '2011-12-01'                        # estimated day of enactment, definitely active in 2012
law_id.ix[40,1] = '1950-01-01'                        # updated in 2013 (originally in 2011) to be stricter but still considered non-strict
law_id.ix[41,1] = '2004-01-01'                        # effective day based on legislative proceedings
law_id.ix[42,[1,2]] = ['1990-01-01', '2012-10-25']    # non-strict: estimated day of enactment, strict: date when it was approved by court
law_id.ix[43,1] = '1997-01-01'                        # estimated date of enactment, became strict in 2014 (still under review as well)
law_id.ix[44,1] = '2009-06-01'                        # estimated date of enactment, bill was discussed in March of that year
law_id.ix[46,[1,2]] = ['1996-01-01','2012-08-20']     # non-strict: estimated date of enactment, strict: day that it was approved by court, could have been enacted later
law_id.ix[47,1] = '2005-06-01'                        # estimated date of enactment, it was passed by state legislature in April 2005

#determe first year that laws affected elections (all relevant elections are in November)
#map to current or next year depending on month of enactment
law_id['nonstrict_firstyear'] = map(lambda x: (x.year + 1) if x.month > 11 else 'NaN' if x.month < 0 else x.year, pd.to_datetime(law_id.non_strict_year))
law_id['strict_firstyear'] = map(lambda x: (x.year + 1) if x.month > 11 else 'NaN' if x.month < 0 else x.year, pd.to_datetime(law_id.strict_year))

#show dataframe
law_id
```

### 1C. Complete Dataframe including Voter Turnout across Ethnicity and Voter ID Law Status

Here we organized all of our ethnic voting and voter id law status data into a single dataframe showing voter turnout across groups for different years and states, in addition to the voter ID law status of each state at the time of the election. Statuses in the table are defined as follows:
    <ul>
    <li> 0: No law in effect for this election
    <li> 1: Non-strict law in effect for this election </li>
    <li> 2: Strict law in effect for this election  </li>

We also clean our data to account for disrepencies between years, among other minor things.

```python
#merge voter ID law dataframe with dataframe on voter turnout across race and state 
racedata = pd.merge(df, law_id, on = ['State'])

#assign each state to a category based on the type of law in effect
racedata['status'] = 2 * (racedata['strict_firstyear']<=racedata['Year']).astype(int) + (racedata['nonstrict_firstyear']<=racedata['Year']).astype(int) 

#change 3's to 2's to make up for the fact that we added states that had both strict and non strict id laws twice
for row in racedata.index:
    if racedata.ix[row,'status'] == 3:
        racedata.ix[row,'status'] = 2

#DATA CLEANING
#Remove periods from beginning of entries in 'Race and Hispanic Origin'
racedata['Race and Hispanic origin'] = racedata['Race and Hispanic origin'].map(lambda x: x.replace('.','')) 

#change non float entries to 'nan'
for row in racedata.index:
    for col in range(2, len(racedata.columns)):
        if racedata.ix[row, col] ==  '(B)' or racedata.ix[row, col] ==  ' (B)' or racedata.ix[row, col] ==  '(B) ':
            racedata.ix[row, col] = float('NaN') 

#correct for discrepancies in ethnicity names 
for i in range(len(racedata)):
    if racedata.ix[i, 'Race and Hispanic origin']=='Non-Hispanic White':
        racedata.ix[i, 'Race and Hispanic origin']='White non-Hispanic alone'
    elif racedata.ix[i, 'Race and Hispanic origin']=='Non-Hispanic Black':
        racedata.ix[i, 'Race and Hispanic origin']='Black alone'
    elif racedata.ix[i, 'Race and Hispanic origin']=='Asian and Pacific Islander':
        racedata.ix[i, 'Race and Hispanic origin']='Asian alone'

#list of all race categories for later use
races = ['Total','White non-Hispanic alone', 'Black alone', 'Asian alone', 'Hispanic (of any race)']

#show head
racedata.head()
```

```python
"""
Function
--------
separator

        Helper-function - Given groupby object for states and list of states, group data by racial group for that list. 

Inputs
------
state : string
    A name of a state in the form seen in racedata dataframe. 
    
groups : groupby object
    Racedata (or similar dataframe) grouped by state. 

Returns
-------
subgroup: list of groups
    List of groups gotten by grouping state group by race.
    
""" 

def separator(state, groups):
    data = groups.get_group(state)
    
    subgroup = [0 for k in range(5)]
    for i,k in zip(range(5),races):
        subgroup[i] = data.groupby('Race and Hispanic origin').get_group(k)

    return subgroup
```

```python
#this function takes in a table where states are distinguished from demographic information by a dot (.) in front of the latter, returning a list of states and a list of tables for each state
def splitbyStateDot(table):
    indices=pd.DataFrame([item[0]=='.' for item in table[table.columns[0]][0:(len(table)-5)]])#trying to find states automatically
    indices=indices.append([False, False, False, False, False]).reset_index(drop=True)
    firstcol=table[table.columns[0]].reset_index(drop=True)
    statesLoc=firstcol[indices[0]==False].index
    return [item for item in firstcol[indices[0]==False]], [table.ix[statesLoc[i]:(statesLoc[i+1]-1), :] for i in range(len(statesLoc)-1)]
```

```python
#this function takes in a table where states are distinguished from demographic information by nan in front of the latter, returning a list of states and a list of tables for each state

def splitbyStateNull(table):
    indices=table[table.columns[0]].isnull()
    indices[len(indices)-5:len(indices)]=False
    firstcol=table[table.columns[0]]
    statesLoc=firstcol[indices==False].index
    return [item for item in firstcol[indices==False]], [table.ix[statesLoc[i]:(statesLoc[i+1]-1), :] for i in range(len(statesLoc)-1)]
```
