```python
import requests, zipfile, StringIO
import pandas as pd
from pandas import DataFrame, read_csv
import math
import numpy as np
```

```python
urllist=['http://www.census.gov/hhes/www/socdemo/voting/publications/p20/2012/Table04b.csv','http://www.census.gov/hhes/www/socdemo/voting/publications/p20/2010/Table4b_2010.csv', 'http://www.census.gov/hhes/www/socdemo/voting/publications/p20/2008/Table%2004b.csv', 'http://www.census.gov/hhes/www/socdemo/voting/publications/p20/2006/tab04b.csv']
```

```python
count=0
for year in [2012, 2010, 2008, 2006]:
    url=urllist[(2012-year)/2]
    table=pd.read_csv(StringIO.StringIO(requests.get(url).content))
    if (year==2012):
        changeCol=dict(zip(table.columns, table.ix[2,:]))#change the columns to match what we want them to be #this portion also not analgous for years
        table.rename(columns=changeCol, inplace=True)
        table=table.drop([0,1,2]).reset_index(drop=True)#ditto this part
        table=table.ix[:,(~pd.isnull(table.columns))] #removing extra nan columns
        stateDict=dict(zip([table.ix[11*i,0] for i in range(52)], [table[11*i:11*(i+1)] for i in range(52)]))#dictionary for info on each state 
    elif year==2010:
        changeCol=dict(zip(table.columns, table.ix[1,:]))#change the columns to match what we want them to be #this portion also not analgous for years
        table.rename(columns=changeCol, inplace=True)
        table=table.drop([0,1]).reset_index(drop=True)#ditto this part
        table=table.ix[:, 0:14]
        table.ix[0, 'STATE']='All'
        for i in range(52):
               stateDict[table.ix[11*i,0]]=[stateDict[table.ix[11*i,0]], table[11*i:11*(i+1)]]
    elif (year==2008)|(year==2006):
        table.ix[4,table.columns[0:3]]=table.ix[3,table.columns[0:3]]
        changeCol=dict(zip(table.columns, table.ix[4,:]))#change the columns to match what we want them to be #this portion also not analgous for years
        table.rename(columns=changeCol, inplace=True)
        table=table.drop([0,1,2,3,4]).reset_index(drop=True)#ditto this part
        table=table.ix[:,(~pd.isnull(table.columns))] #removing extra nan columns
        table.ix[0, 'State, sex, race, and Hispanic origin']='All'
        for i in range(52):
            existinglist=stateDict[table.ix[12*i,0]]
            existinglist.append(table[12*i:12*(i+1)])
            stateDict[table.ix[12*i,0]]=existinglist 
        
```
