```python
import requests, zipfile, StringIO
import pandas as pd
from pandas import DataFrame, read_csv
import math
import numpy as np
```

```python
url='http://www.census.gov/hhes/www/socdemo/voting/publications/p20/2012/Table04b.csv' #url for sex and race
table=pd.read_csv(StringIO.StringIO(requests.get(url).content)) #read in CSV
```

```python
changeCol=dict(zip(table.columns, table.ix[2,:]))#change the columns to match what we want them to be
table.rename(columns=changeCol, inplace=True)
table=table.drop([0,1,2]).reset_index(drop=True)
table=table.ix[:,(~pd.isnull(table.columns))] #removing extra nan columns
```

```python
#Note: I found how many non-nan entries there were with (~pd.isnull(table['State'])).sum()
stateDict=dict(zip([table.ix[11*i,0] for i in range(52)], [table[11*i:11*(i+1)] for i in range(52)]))#dictionary for info on each state 
stateDict['ARKANSAS']
```
